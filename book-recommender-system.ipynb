```python
import numpy as np
import pandas as pd
```

```python
books=pd.read_csv("books.csv")
users=pd.read_csv("users.csv")
ratings=pd.read_csv("ratings.csv")
```

```python
books
```

```python
users.duplicated().sum()
```

```python
ratings_with_name=ratings.merge(books,on="ISBN")
```

```python
ratings_with_name.head(2)
```

```python
num_rating_df=ratings_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
```

```python
num_rating_df.rename(columns={"Book-Rating":"num-ratings"},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_rating_df.rename(columns={"Book-Rating":"avg-ratings"},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on="Book-Title")
popular_df
```

```python
popular_df=popular_df[popular_df["num-ratings"]>=250].sort_values("avg-ratings",ascending=False)
popular_df
```

```python
popular_df=popular_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")
```

```python
popular_df[["Book-Title","Book-Author","Image-URL-M","num-ratings","avg-ratings"]]
```

```python
x=ratings_with_name.groupby("User-ID").count()["Book-Rating"]>200
new_users=x[x].index
new_users
```

```python
filtered_rating=ratings_with_name[ratings_with_name["User-ID"].isin(new_users)]
filtered_rating
```

```python
y=filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
famous_books=y[y].index
famous_books
```

```python
final_ratings=filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
final_ratings
```

```python
pt=final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score=cosine_similarity(pt)
similarity_score
```

```python
def recommend(book):
    index=np.where(pt.index==book)[0][0]
    distances=similarity_score[index]
    similar_items=sorted(list(enumerate(distances)),key=lambda x:x[1],reverse=True)[1:11]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books["Book-Title"]==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))
        data.append(item)
    return data
```

```python
recommend("1984")
```
